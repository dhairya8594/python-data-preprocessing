#opening file

```python
with open("initialdata.txt" , encoding = "utf-8")as f:
    data = f.read()

data    
```

```python
#Creating chunks
```

```python
chunks = data.split("\n\n")
chunks = [c for c in chunks if len(c)>3]
chunks[4]
```

```python
def parse_chunks (chunks):
    chunks = chunks.strip()
    sep_chunk = chunks.split("\n")
    username = sep_chunk[0]
    no_of_post = int(sep_chunk[1].split("post")[0].replace(",",""))
    
    no_of_followers = float(sep_chunk[2].split("followers")[0].replace(",","").replace("K","").replace("M",""))
    if ("K"in sep_chunk[2]):
        no_of_followers = int(no_of_followers*1000)
    elif ("M"in sep_chunk[2]):
        no_of_followers = int(no_of_followers*1000000)
    else :
        no_of_follower = int(no_of_followers)
        
    no_of_following = float(sep_chunk[3].split("following")[0].replace(",","").replace("K","").replace("M",""))
    if ("K"in sep_chunk[3]):
        no_of_following = int(no_of_following*1000)
    elif ("M"in sep_chunk[3]):
        no_of_following = int(no_of_following*1000000)
    else :
        no_of_following = int(no_of_following)
    
    name = sep_chunk[4]
    if(len(sep_chunk)>5):
        catagories = sep_chunk[5]
        bio = "\n".join(sep_chunk[6:])
    else :
        catagories = "Unkown"
        bio = ""
        
    return {"username": username, "no_of_post": no_of_post, "no_of_followers": no_of_followers,"no_of_following": no_of_following, "name": name, "catagories": catagories, "bio": bio}

parse_chunks(chunks[1])   
```

```python
all_chunks = []
for chunk in chunks:
    parsed_chunks = parse_chunks(chunk)
    all_chunks.append(parsed_chunks)
```

```python
import json
s = json.dumps(all_chunks, indent=4)
with open("data.json","w")as f:
    f.write(s)
```

## Who has the maximum followers?
